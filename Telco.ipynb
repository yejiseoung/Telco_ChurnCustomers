### How many customers decided to end the services?

Let's figure out how many customers churn or retain, which is my main interest

```python
x = df.groupby(['Churn'])['Churn'].count()
y = len(df)
r = ((x/y)).round(2)
ratio = pd.DataFrame(r).T

fig, ax = plt.subplots(1,1,figsize=(6.5, 2), dpi=150)
background_color = '#fbfbfb'
fig.patch.set_facecolor(background_color)
ax.set_facecolor(background_color)

ax.barh(ratio.index, ratio['No'], color='gray', alpha=0.9, 
       ec=background_color, label='Retain')
ax.barh(ratio.index, ratio['Yes'], left=ratio['No'], color='#0e4f66',
       alpha=0.9, ec=background_color, label='Churn')

ax.set_xlim(0,1)
ax.set_xticks([])
ax.set_yticks([])
ax.legend().set_visible(False)
for s in ['top', 'left', 'right', 'bottom']:
    ax.spines[s].set_visible(False)

for i in ratio.index:
    ax.annotate(f"{int(ratio['No'][i]*100)}%", xy=(ratio['No'][i]/2, i),
               va = 'center', ha= 'center', fontsize=32, fontweight='light',
               fontfamily='serif', color='white')
    ax.annotate('Retain', xy=(ratio['No'][i]/2, -0.25), va='center', ha='center',
               fontsize=12, fontweight='light', fontfamily='serif', color='white')
    
for i in ratio.index:
    ax.annotate(f"{int(ratio['Yes'][i]*100)}%", xy=(ratio['No'][i]+ratio['Yes'][i]/2, i),
               va='center', ha='center', fontsize=32, fontweight='light', fontfamily='serif', color='white')
    ax.annotate('Churn', xy=(ratio['No'][i]+ratio['Yes'][i]/2, -0.25),
               va='center', ha='center', fontsize=12, fontweight='light', fontfamily='serif', color='white')

fig.text(0.125, 0.95, 'How many customers have decided to end the services?', fontfamily='serif',
        fontsize=15, fontweight='bold')


plt.show()
```

73 percent of customers retain their services, while 27 percent of those churned.

```python
def barPerc(df, xVar, ax):
    """ Add percentage for hues to bar plots
    args:
        df: pandas dataframe
        xVar (string): X variable
        ax: Axes object
    """
    # 1. How many X categories
    ## check for Nan and Remove
    numX = len([x for x in df[xVar].unique() if x==x])
    
    # 2. The bars are created in hue order, organize them
    bars = ax.patches
    
    ## 2a. For eaxh X variable
    for ind in range(numX):
        ## 2b. Get hue bar
        hueBars = bars[ind:][::numX]
        
        ## 2c. Get the total height (for percentages)
        total = sum([x.get_height() for x in hueBars])
        
        # 3. Print the percentage on the bars
        for bar in hueBars:
            ax.text(bar.get_x() + bar.get_width()/2.,
                   bar.get_height() - 0.1 * bar.get_height(),
                   f'{bar.get_height()/total:.2%}',
                   ha='center', va='top', color='#f6f6f6')
            
def GrpSubplots(df, xVar, ax, axis=None, ticklabels=None):
    """ Create subplots based on X variables"""
    df.groupby([xVar, 'Churn']).size().unstack('Churn').apply(lambda x: x*100/x.sum(), axis=axis).plot.bar(rot=0,
                                                                                                           width=0.9,
                                                                                                           color=['grey', '#0e4f66'],
                                                                                                           ax=ax);
    ax.set_title(f'{xVar}'.title(), fontsize=14, fontweight='bold', fontfamily='serif', color='#323232', loc='left')
    ax.grid(color='gray', linestyle=':', axis='y', zorder=0, dashes=(1,5))
    ax.set_xticklabels(ticklabels)
    ax.yaxis.set_major_formatter(mtick.PercentFormatter())
    ax.yaxis.set_major_locator(mtick.MultipleLocator(10))
    ax.get_legend().remove()
    
    barPerc(df, xVar, ax)
```

```python
background_color = '#fbfbfb'

fig = plt.figure(figsize=(20,10), dpi=150)
fig.patch.set_facecolor(background_color)
gs = fig.add_gridspec(2,2)
gs.update(wspace=0.35, hspace=0.27)

# axes as a list
axes = [fig.add_subplot(gs[0,0]),
        fig.add_subplot(gs[0,1]),
        fig.add_subplot(gs[1,0]),
        fig.add_subplot(gs[1,1])]

# categories as a list
cats = ['gender', 'SeniorCitizen', 'Partner', 'Dependents']

# ticklabels as a list
tlabs = [['Male', 'Female']] + [['No Senior', 'Senior']] + [['No', 'Yes']]*2

axes[0].text(-0.80, 125, 'Retain', fontsize=24, 
        fontweight='bold', fontfamily='serif', color='gray')
axes[0].text(-0.05, 125, 'vs.', fontsize=24,
        fontweight='bold', fontfamily='serif', color='#323232')
axes[0].text(0.35, 125, 'Churn', fontsize=24,
        fontweight='bold', fontfamily='serif', color='#0e4f66')
axes[0].text(-0.80, 113, '- The gender proportions in both Retain and Churn groups are similar.',
        fontsize=20, fontfamily='serif', color='#323232')
axes[0].text(-0.80, 105, '- Senior group is more likely to end the services compared to non-senior group.',
        fontsize=20, fontfamily='serif', color='#323232')
axes[0].text(-0.80, 97, '- When customers have NO partner, they are more likely to end the services compared to those having partner. ',
        fontsize=20, fontfamily='serif', color='#323232')
axes[0].text(-0.80, 89, '- When customers have NO dependents, they are more likely to end the services compared to those having dependents.',
        fontsize=20, fontfamily='serif', color='#323232')

for ax, xVar, ticklabels in zip(axes, cats, tlabs):
    GrpSubplots(df, xVar, ax, axis=1, ticklabels=ticklabels)
```

## Conclusion

In this project, we investigated what features have an effect on the customer retent. 

We found that customers who were using fiber optic for Internet service tend to end the service compared to those using DSL. We cannot conclude that Fiber optic for Internet Service might have poor service compared to DSL from this dataset, but in the further research, if we can get more information on like service satisfaction, Internet service quality, the price of those services, then we can find more why customers churned. 

We also found that churned customers tended to be charged monthly more compared to retaining customers, and high monthly payment can be an effect on customers decision on retaining or not. 

Churned customers are distributed on relatively new contract on the service compared to retained customers. Since they have less loyalty to the service, they easily decided to quit the service. 

Interestingly, when customers contract on month-to-month, they are more likely to churn compared to those contract on year-contract.
